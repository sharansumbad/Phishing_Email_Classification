```python
#function to get the frequency of stop words
def stop_freq(string):
    cnt = 0
    stop_words = set(stopwords.words('english'))
    for word in string.split():
        if word in stop_words:
            cnt += 1
        else:
            cnt = cnt 
    return cnt 

#function to get the frequency of top 5 words in legit emails
def cnt_top5_legit(string):
    cnt = 0
    top5_words = ['TRUMP' , 'MAY' , 'PM' , 'REPUBLICAN' , 'STATE']
    for word in string.upper().split():
        if word in top5_words:
            cnt += 1
        else:
            cnt = cnt 
    return cnt 

#function to get the frequency of top 5 words in legit emails
def cnt_top5_phish(string):
    cnt = 0
    top5_words = ['ACCOUNT' , 'EBAY' , 'PAYPAL' , 'SECURITY' , 'BANK']
    for word in string.upper().split():
        if word in top5_words:
            cnt += 1
        else:
            cnt = cnt 
    return cnt 
    
#function to get the frequency of uppercase letters
def count_upper_case_letters(str_obj):
    count = 0
    for elem in str_obj:
        if elem.isupper():
            count += 1
    return count

#function to get the frequency of punctuations

def count_punc(st):
    count = lambda l1, l2: len(list(filter(lambda c: c in l2, l1)))
    cnt_punct = count(st, string.punctuation)
    return cnt_punct

# function to process the text data
def process_text(text):
    no_punc = [char for char in text if char not in string.punctuation]
    no_punc = ''.join(no_punc)
    return ' '.join([word for word in no_punc.split() if word.lower() not in stopwords.words('english')])

#function to remove the numeric characters
def remove_numric_character(x):
    x = re.sub(r'\b[0-9]+\b\s*', '', x)
    return x   


    
```

## Modeling 

```python
# importing libraries for modeling
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
import pandas as pd
import numpy as np

# cleaing to remove nan and other unsuitable values
def clean_dataset(df):
    assert isinstance(df, pd.DataFrame), "df needs to be a pd.DataFrame"
    df.dropna(inplace=True)
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(1)
    return df[indices_to_keep].astype(np.float64)
```

```python
preped_data = clean_dataset(preped_data)
```

```python
preped_data.info()
```

```python
# defining the targert varaible and features
X = preped_data.loc[:, preped_data.columns!='Phish']
Y = preped_data['Phish']
```

```python
#train test split
train_x,test_x,train_y,test_y = train_test_split(X,Y,test_size = 0.25)
```

```python
## Naive Bayes Model training and testing


mnb = MultinomialNB(alpha=1.9) # alpha by default is 1. alpha must always be > 0. 
# alpha is the '1' in the formula for Laplace Smoothing (P(words))
mnb.fit(train_x,train_y)
y_pred1 = mnb.predict(test_x)
print("Accuracy Score for Naive Bayes : ", accuracy_score(y_pred1,test_y))
```

```python
# evaluation metrics 

from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(test_y, y_pred1, target_names= ['Phish','Legit' ]))
```

```python
rfc = RandomForestClassifier(n_estimators=100,criterion='gini')
# n_estimators = No. of trees in the forest
# criterion = basis of making the decision tree split, either on gini impurity('gini'), or on infromation gain('entropy')
rfc.fit(train_x,train_y)
y_pred3 = rfc.predict(test_x)
print("Accuracy Score of Random Forest Classifier : ", accuracy_score(y_pred3,test_y))
```
